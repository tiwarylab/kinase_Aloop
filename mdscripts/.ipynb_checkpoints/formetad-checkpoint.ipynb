```python
def classify_dfg_state(distance1, distance2):
    """
    Classify DFG state based on Dunbrack distances using specific thresholds.
    
    Parameters:
    - distance1, distance2: The two distances to be compared.
    
    Returns:
    - DFG state classification.
    """
    
    if distance1 < 11:
        if distance2 > 11:
            return "DFG-in"
        else:
            return "DFG-inter"
        
    else:
        if distance2 <= 14:
            return "DFG-out"
        else:
            return "DFG-inter"
        
        return "DFG-out"

def classify_trajectory(trajectory):
    return [classify_dfg_state(row[0], row[1]) for row in trajectory]

def plot_trajectory(states):
    # Convert states to numbers for plotting
    state_map = {"DFG-in": 1, "DFG-inter": 2, "DFG-out": 3, "unassigned": 4}
    numerical_states = [state_map[state] for state in states]
    
    plt.figure(figsize=(10, 5))
    plt.plot(numerical_states, marker='o', linestyle='-')
    
    # Set y-axis labels
    plt.yticks(list(state_map.values()), list(state_map.keys()))
    
    plt.xlabel('Index')
    plt.ylabel('State')
    plt.title('DFG State Trajectory')
    plt.grid(True, which='both', linestyle='--', linewidth=0.5)
    plt.tight_layout()
    plt.show()

# Test
trajectory = dists[0][0][::10,-3:-1]*10
states = classify_trajectory(trajectory)
plot_trajectory(states)
```

```python
import mdtraj as md
import py3Dmol
import tempfile
import mdtraj as md
import py3Dmol
import tempfile
import os

def translate_mdtraj_frame(traj_frame, offset):
    """Translates an mdtraj frame."""
    traj_frame.xyz[0] += offset
    return traj_frame

def traj_to_pdb_string(traj):
    with tempfile.NamedTemporaryFile(suffix=".pdb", delete=False) as temp_file:
        traj.save_pdb(temp_file.name)
        
        with open(temp_file.name, 'r') as f:
            pdb_string = f.read()

    os.remove(temp_file.name)
    return pdb_string

def view_structures(xtc_path, pdb_path, frame_indices):
    # Load only the specified frames from the xtc file using the pdb as a topology
    traj = md.load(pdb_path)
    
    # Select only the protein atoms
    protein_indices = traj.topology.select('protein')
    
    # Extract the protein topology
    topology = traj.topology.subset(protein_indices)
    traj_frames = [md.load_xtc(xtc_path, top=topology, frame=i) for i in frame_indices]
    traj = md.join(traj_frames)
    reference_frame = traj[0]
    backbone_indices = traj.topology.select('backbone and protein')
    
    traj.superpose(reference_frame, atom_indices=backbone_indices)
        # Create a py3Dmol view object
    view = py3Dmol.view(width=800, height=400)  # Adjust width based on number of frames

    xOffset = 0  # Starting offset for x-axis translation

    for frame in traj:
        # Convert the frame to PDB string format
        #frame = translate_mdtraj_frame(frame, [xOffset, 0, 0])
        pdb_string = traj_to_pdb_string(frame)

        # Add the model to the view
        model = view.addModel(pdb_string, 'pdb')

        # Apply the visualization specifications
        
        model.setStyle({}, {'cartoon': {'color': 'grey'}}) # entire structure in grey cartoon
        
        model.setStyle({'resi': 186}, {'cartoon': {'color': '#006400'}}) # resid 186 in dark green (hex code for darkgreen)
        model.setStyle({'resi': 187}, {'cartoon': {'color': '#00008B'}})  # resid 187 in dark blue (hex code for darkblue)
        model.setStyle({'resi': 188}, {'cartoon': {'color': '#4B0082'}})  # resid 188 in indigo (a close substitute for darkpurple)
        model.addStyle({'resi': 186}, {'stick': {'color': '#006400','radius': 0.4}})  # resid 186 in dark green
        model.addStyle({'resi': 187}, {'stick': {'color': '#00008B','radius': 0.4}})  # resid 187 in dark blue
        model.addStyle({'resi': 188}, {'stick': {'color': '#4B0082','radius': 0.4}})  # resid 188 in indigo
        


        # Translate the structure along the x-axis
    
#        xOffset += 5  # Increase offset for the next structure

    view.zoomTo()
    return view

xtc_file = dirpath+"pmeta/WT/dunbrack_metad_0/prod_0.xtc"
pdb_file = dirpath+'equilruns/WT/equil_0/restrainedNPT_0.pdb'

indices = [10,50000,75000,100000]  # Replace with desired frame indices

view = view_structures(xtc_file, pdb_file, indices)

view.show()
```

```python
import tempfile
import os
import mdtraj as md
import py3Dmol
from matplotlib.cm import get_cmap

def traj_to_pdb_string(traj):
    with tempfile.NamedTemporaryFile(suffix=".pdb", delete=False) as temp_file:
        traj.save_pdb(temp_file.name)
        with open(temp_file.name, 'r') as f:
            pdb_string = f.read()
    os.remove(temp_file.name)
    return pdb_string

def view_structures(xtc_path, pdb_path, frame_indices):
    traj = md.load(pdb_path)
    protein_indices = traj.topology.select('protein')
    topology = traj.topology.subset(protein_indices)
    traj_frames = [md.load_xtc(xtc_path, top=topology, frame=i) for i in frame_indices]
    traj = md.join(traj_frames)
    reference_frame = traj[0]
    backbone_indices = traj.topology.select('backbone and protein')
    traj.superpose(reference_frame, atom_indices=backbone_indices)
    
    # Generate a color map
    plasma_cmap = get_cmap('plasma')
    num_frames = len(frame_indices)
    colors = [plasma_cmap(i / (num_frames - 1)) for i in range(num_frames)]
    colors = ['#%02x%02x%02x' % (int(r*255), int(g*255), int(b*255)) for r, g, b, _ in colors]

    view = py3Dmol.view(width=800, height=400)

    for i, frame in enumerate(traj):
        pdb_string = traj_to_pdb_string(frame)
        model = view.addModel(pdb_string, 'pdb', model=i)
        color = colors[i]
        model.setStyle({}, {'cartoon': {'color': color}}, model=i)
        model.setStyle({'resi': '186 or 187 or 188'}, {'cartoon': {'color': color}}, model=i)
        model.setStyle({}, {'stick': {'color': color, 'radius': 0.4}}, model=i)
        
    view.zoomTo()
    return view

dirpath="/home/bvani/scratch/kinase/"

xtc_file = dirpath + "pmeta/WT/dunbrack_metad_0/prod_0.xtc"
pdb_file = dirpath + 'equilruns/WT/equil_0/restrainedNPT_0.pdb'
indices = [10, 50000, 75000, 100000]

view = view_structures(xtc_file, pdb_file, indices)
view.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

import os

def find_bias_files(path):
    # List all files in the directory
    files = os.listdir(path)
    
    # Filter the list for files that start with "bias" and end with ".npy"
    bias_files = [f for f in files if f.startswith('bias') and f.endswith('.npy')]
    
    return bias_files
```

```python
# Loading the numpy arrays
file_names = [dirpath+"pmeta/"+p+"/dunbrack_metad_0/"+find_bias_files(dirpath+"pmeta/"+p+"/dunbrack_metad_0/")[0] for p in types]

# Function to load arrays and preprocess them
def load_and_preprocess(file_names):
    arrays = [np.load(f) for f in file_names]

    # Multiply each array by -1
    arrays = [-1 * array for array in arrays]

    # Offset each array to have the same minimum value
    global_min = min(array.min() for array in arrays)
    arrays = [array - (array.min() - global_min) for array in arrays]

    return arrays

# Generate x and y values based on array shapes
def generate_xy_edges(array, x_edges=(0.25, 2.5), y_edges=(0.25, 2.5)):
    y, x = np.mgrid[y_edges[0]:y_edges[1]:array.shape[0]*1j, 
                    x_edges[0]:x_edges[1]:array.shape[1]*1j]
    return x, y

# List of file names
arrays = load_and_preprocess(file_names)

# Setting up the figure
fig, axes = plt.subplots(2, 2, figsize=(10, 10))
cbar_ax = fig.add_axes([0.92, 0.15, 0.02, 0.7])  # Adjust for colorbar

# Determine global min and max for consistent color scales
vmin = min(array.min() for array in arrays)
vmax = max(array.max() for array in arrays)

# Create the contour plots in a 2x2 arrangement
for ax, array in zip(axes.flat, arrays):
    X, Y = generate_xy_edges(array)
    c = ax.contourf(X, Y, array, cmap="plasma", vmin=vmin, vmax=vmax, levels=100)  # Smooth gradient
    ax.contour(X, Y, array, levels=10, colors='white')  # 10 contour lines
    ax.set_xlim(0.25, 2.5)
    ax.set_ylim(0.25, 2.5)

# Adding a single colorbar to the right
fig.colorbar(c, cax=cbar_ax)

plt.tight_layout()
plt.subplots_adjust(right=0.9)  # Space for the colorbar
plt.show()
```
